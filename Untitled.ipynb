```python
import json
import pandas as pd

# Specify the path to your JSON file
file_path = 'data.json'

# Open the file and load JSON data
with open(file_path, 'r') as file:
    data = json.load(file)

# Now, 'data' contains the content of your JSON file without metadata

# Convert JSON data to a pandas DataFrame
df = pd.DataFrame(data)

# Now, 'df' is a DataFrame containing the JSON data
print(df)
```
